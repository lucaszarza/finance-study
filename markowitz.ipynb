## 7. Results Analysis

Let's analyze our optimized portfolios:

```python
def display_portfolio_results(portfolio_result, portfolio_name, tickers):
    """
    Display portfolio optimization results in a nice format
    """
    if portfolio_result is None:
        print(f"❌ {portfolio_name}: Not calculated yet")
        return
        
    print(f"\n📊 {portfolio_name}:")
    print(f"   Expected Return: {portfolio_result['return']:.2%}")
    print(f"   Volatility: {portfolio_result['volatility']:.2%}")
    if 'sharpe' in portfolio_result:
        print(f"   Sharpe Ratio: {portfolio_result['sharpe']:.3f}")
    
    print("   Asset Allocation:")
    for ticker, weight in zip(tickers, portfolio_result['weights']):
        if weight > 0.01:  # Only show weights > 1%
            print(f"   {ticker}: {weight:.1%}")

# Display results
print("=== PORTFOLIO OPTIMIZATION RESULTS ===")

display_portfolio_results(min_vol_result, "Minimum Variance Portfolio", TICKERS)
display_portfolio_results(max_sharpe_result, "Maximum Sharpe Ratio Portfolio", TICKERS)

# Compare with individual stocks
print("\n📈 Individual Stock Performance:")
individual_volatilities = np.sqrt(np.diag(cov_matrix))
for ticker, ret, vol in zip(TICKERS, mean_returns, individual_volatilities):
    sharpe = (ret - 0.02) / vol
    print(f"   {ticker}: Return {ret:.2%}, Volatility {vol:.2%}, Sharpe {sharpe:.3f}")
```
